# 06 — Structural Analysis: PCA, Clustering and Namibia Benchmarking

This notebook implements the analytical framework described in Chapter 3.

Steps:
1. Construct country structural profiles
2. Standardize indicators
3. Compute structural similarity (distance to Namibia)
4. Perform Principal Component Analysis (PCA)
5. Identify structural peer economies
6. Apply K-means clustering
7. Construct Namibia Benchmark Index (NBI)

Input:
- final_analysis_dataset.csv

Output:
- Structural comparison figures
- Cluster typology
- Namibia Benchmark Index

##LOAD FINAL ANALYTICAL DATASET

```python
df = pd.read_csv(
    OUTPUT_DATA + "final_analysis_dataset.csv"
)

df.head()
```

##CONSTRUCT COUNTRY STRUCTURAL PROFILES

```python
benchmark_vars = [
    "Oil_Rents_GDP",
    "GDP_per_capita",
    "GDP_growth",
    "Inflation",
    "Gov_Debt_GDP",
    "Control of Corruption",
    "Government Effectiveness",
    "Rule of Law"
]

country_avg = (
    df.groupby(["Country Code","Country Name"])
      [benchmark_vars]
      .mean(numeric_only=True)
      .reset_index()
)
```

##VARIABLE STANDARDIZATION

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

country_avg[benchmark_vars] = scaler.fit_transform(
    country_avg[benchmark_vars]
)
```

##STRUCTURAL SIMILARITY (DISTANCE TO NAMIBIA)

```python
from scipy.spatial.distance import cdist

namibia_vector = country_avg[
    country_avg["Country Code"]=="NAM"
][benchmark_vars]

country_avg["distance_to_namibia"] = cdist(
    country_avg[benchmark_vars],
    namibia_vector,
    metric="euclidean"
)
```

##PRINCIPAL COMPONENT ANALYSIS (PCA)

```python
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

X = country_avg[benchmark_vars]

imputer = SimpleImputer(strategy="median")
X = imputer.fit_transform(X)

pca = PCA(n_components=2)

components = pca.fit_transform(X)

country_avg["PC1"] = components[:,0]
country_avg["PC2"] = components[:,1]

pca.explained_variance_ratio_
```

##STRUCTURAL SIMILARITY MAP

```python
plt.figure(figsize=(10,8))

plt.scatter(country_avg["PC1"], country_avg["PC2"], alpha=0.6)

namibia = country_avg["Country Code"]=="NAM"

plt.scatter(
    country_avg.loc[namibia,"PC1"],
    country_avg.loc[namibia,"PC2"],
    s=200,
    marker="*"
)

plt.title("Structural Similarity Map of Oil Economies")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
save_fig("figure_pca_structural_map.png")
```


##PCA Interpretation — Component Loadings


This table shows how each economic and governance variable contributes
to the principal components. Loadings are used to interpret the
economic meaning of PC1 and PC2.

```python
# PCA Loadings
loadings = pd.DataFrame(
    pca.components_.T,
    columns=["PC1","PC2"],
    index=benchmark_vars
)

loadings.round(3)
```

##IDENTIFY STRUCTURAL PEERS (DISTANCE-BASED)

```python
peers = country_avg.sort_values(
    "distance_to_namibia"
).iloc[1:6]

peers
```

##PEER COMPARISON BAR GRAPH

```python
comparison = country_avg[
    country_avg["Country Code"].isin(
        list(peers["Country Code"]) + ["NAM"]
    )
]

comparison.set_index("Country Name")[benchmark_vars].T.plot(
    kind="bar",
    figsize=(12,6)
)

plt.title("Namibia vs Structural Peer Economies")
plt.ylabel("Standardized Value")
plt.xticks(rotation=45)
plt.show()
save_fig("figure_peer_comparison_bar.png")
```

## Namibia Structural Evolution (Early vs Late Period)

```python
# Variables used in benchmarking
features = [
    "Oil_Rents_GDP",
    "GDP_per_capita",
    "GDP_growth",
    "Inflation",
    "Gov_Debt_GDP",
    "Control of Corruption",
    "Government Effectiveness",
    "Rule of Law"
]

# Split dataset
early_df = df[df["Year"] <= 2010]
late_df  = df[df["Year"] > 2010]

# Compute country averages
early_avg = early_df.groupby("Country Code")[features].mean()
late_avg  = late_df.groupby("Country Code")[features].mean()

# Extract Namibia only
namibia_early = early_avg.loc["NAM"]
namibia_late  = late_avg.loc["NAM"]

print("Namibia early averages:")
print(namibia_early)

print("\nNamibia late averages:")
print(namibia_late)
```

```python
# =====================================
# Namibia Structural Evolution
# (Relative to Global Distribution)
# =====================================

import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------------
# Variables used in analysis
# -------------------------------------
features = [
    "Oil_Rents_GDP",
    "GDP_per_capita",
    "GDP_growth",
    "Inflation",
    "Gov_Debt_GDP",
    "Control of Corruption",
    "Government Effectiveness",
    "Rule of Law"
]

# -------------------------------------
# Split dataset into early vs late
# -------------------------------------
early_df = df[df["Year"] <= 2010]
late_df  = df[df["Year"] > 2010]

# -------------------------------------
# Compute country averages
# -------------------------------------
early_avg = early_df.groupby("Country Code")[features].mean()
late_avg  = late_df.groupby("Country Code")[features].mean()

# -------------------------------------
# Extract Namibia values
# -------------------------------------
namibia_early = early_avg.loc["NAM"]
namibia_late  = late_avg.loc["NAM"]

# -------------------------------------
# Combine into comparison dataframe
# -------------------------------------
comparison = pd.DataFrame({
    "Early": namibia_early,
    "Late": namibia_late
}).T

# -------------------------------------
# IMPORTANT:
# Use SAME scaler from PCA stage
# (DO NOT refit scaler)
# -------------------------------------
comparison_std = pd.DataFrame(
    scaler.transform(comparison),
    columns=features,
    index=["Early Period", "Late Period"]
)

# -------------------------------------
# Plot
# -------------------------------------
comparison_std.T.plot(
    kind="bar",
    figsize=(12,6)
)

plt.title("Structural Evolution of Namibia (Relative to Global Distribution)")
plt.ylabel("Standardized Value (Global Z-score)")
plt.axhline(0, linestyle="--", linewidth=1)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##K-MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=4, random_state=42)

country_avg["cluster"] = kmeans.fit_predict(
    country_avg[["PC1","PC2"]]
)

silhouette_score(
    country_avg[["PC1","PC2"]],
    country_avg["cluster"]
)
```


## Cluster Validation — Choosing Number of Clusters

Silhouette scores are evaluated across multiple cluster sizes
to confirm that the selected four-cluster solution provides
meaningful structural separation.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

scores = []

for k in range(2,7):
    km = KMeans(n_clusters=k, random_state=42)
    labels = km.fit_predict(country_avg[["PC1","PC2"]])
    score = silhouette_score(country_avg[["PC1","PC2"]], labels)
    scores.append(score)

plt.figure(figsize=(6,4))
plt.plot(range(2,7), scores, marker="o")
plt.title("Silhouette Score by Number of Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.show()
```

##CLUSTER STRUCTURAL SUMMARY

```python
cluster_summary = (
    country_avg
    .groupby("cluster")[benchmark_vars]
    .mean()
)

cluster_summary.round(2)
```

## Cluster Structural Profiles

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.heatmap(
    cluster_summary,
    annot=True,
    cmap="coolwarm",
    center=0
)

plt.title("Structural Characteristics of Resource Economy Clusters")
plt.ylabel("Cluster")
plt.xlabel("Indicators")

plt.show()
```

##NAMIBIA BENCHMARK INDEX (NBI)

```python
frontier = country_avg[
    country_avg["Country Code"]=="NOR"
][benchmark_vars].iloc[0]

namibia = country_avg[
    country_avg["Country Code"]=="NAM"
][benchmark_vars].iloc[0]

institutional_vars = [
    "Control of Corruption",
    "Government Effectiveness",
    "Rule of Law"
]

economic_vars = [
    "GDP_per_capita",
    "GDP_growth",
    "Inflation",
    "Gov_Debt_GDP",
    "Oil_Rents_GDP"
]

institutional_gap = abs(
    namibia[institutional_vars] - frontier[institutional_vars]
).mean()

economic_gap = abs(
    namibia[economic_vars] - frontier[economic_vars]
).mean()

NBI = (institutional_gap + economic_gap) / 2

NBI
```

##RADAR BENCHMARK VISUALIZATION

```python
import numpy as np
import matplotlib.pyplot as plt

# Variables used (UNCHANGED)
vars_used = [
    "Oil_Rents_GDP",
    "GDP_per_capita",
    "GDP_growth",
    "Inflation",
    "Gov_Debt_GDP",
    "Control of Corruption",
    "Government Effectiveness",
    "Rule of Law"
]

# Axis labels
labels = [
    "Oil Rents",
    "Income",
    "Growth",
    "Inflation",
    "Debt",
    "Corruption",
    "Gov Effect.",
    "Rule Law"
]

# Namibia profile
namibia_profile = country_avg[
    country_avg["Country Code"] == "NAM"
][vars_used].iloc[0]

# Frontier (Norway)
frontier_profile = country_avg[
    country_avg["Country Code"] == "NOR"
][vars_used].iloc[0]

# Namibia peer cluster average (UNCHANGED LOGIC)
nam_cluster = country_avg.loc[
    country_avg["Country Code"] == "NAM", "cluster"
].iloc[0]

peer_profile = (
    country_avg[country_avg["cluster"] == nam_cluster][vars_used]
    .mean()
)

# ---------- Radar Geometry ----------
N = len(labels)

angles = np.linspace(0, 2*np.pi, N, endpoint=False)
angles = np.concatenate([angles, [angles[0]]])

def close_loop(values):
    return np.concatenate([values, [values[0]]])

namibia_values = close_loop(namibia_profile.values)
frontier_values = close_loop(frontier_profile.values)
peer_values = close_loop(peer_profile.values)

# ---------- Plot ----------
plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)

# Namibia
ax.plot(angles, namibia_values,
        linewidth=3,
        label="Namibia")
ax.fill(angles, namibia_values, alpha=0.15)

# Frontier
ax.plot(angles, frontier_values,
        linestyle="--",
        linewidth=2,
        label="Frontier (Norway)")

# Peer average
ax.plot(angles, peer_values,
        linewidth=2,
        label="Peer Average")

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=12)

plt.title(
    "Namibia Structural Benchmark vs Frontier and Peers",
    size=16,
    pad=25
)

plt.legend(loc="upper right", bbox_to_anchor=(1.25,1.1))
plt.tight_layout()
plt.show()
```

## Namibia Benchmark Gap Decomposition

```python
features = [
    "Oil_Rents_GDP",
    "GDP_per_capita",
    "GDP_growth",
    "Inflation",
    "Gov_Debt_GDP",
    "Control of Corruption",
    "Government Effectiveness",
    "Rule of Law"
]

gap_values = abs(namibia[features] - frontier)

gap_values.sort_values().plot(
    kind="barh",
    figsize=(8,6)
)

plt.title("Namibia Distance from Frontier by Indicator")
plt.xlabel("Absolute Standardized Gap")

plt.show()
```
