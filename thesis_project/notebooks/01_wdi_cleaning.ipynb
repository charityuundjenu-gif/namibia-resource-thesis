## Load ISO Country Codes
Filter datasets to sovereign countries only.

```python
iso_url = "https://raw.githubusercontent.com/datasets/country-codes/master/data/country-codes.csv"
iso = pd.read_csv(iso_url)

valid_codes = iso["ISO3166-1-Alpha-3"].dropna().unique()

print("ISO codes loaded:", len(valid_codes))
```

## WDI Cleaning Function
Converts raw WDI files into panel (country-year) format.

```python
def clean_wdi(file_path, iso_codes, value_name):

    df = pd.read_excel(file_path, skiprows=3)

    df = df[df["Country Code"].isin(iso_codes)]

    year_cols = [c for c in df.columns if str(c).isdigit()]

    df_panel = df.melt(
        id_vars=["Country Name", "Country Code"],
        value_vars=year_cols,
        var_name="Year",
        value_name=value_name
    )

    df_panel["Year"] = df_panel["Year"].astype(int)
    df_panel[value_name] = pd.to_numeric(
        df_panel[value_name], errors="coerce"
    )

    df_panel = df_panel.dropna(subset=[value_name])

    return df_panel.sort_values(
        ["Country Code", "Year"]
    ).reset_index(drop=True)
```

## Oil Rents (% of GDP)

```python
oil_rents = clean_wdi(
    RAW_DATA + "WDI_OilRents_GDP_raw.xls",
    valid_codes,
    "Oil_Rents_GDP"
)

oil_rents["log_oil_rents"] = np.log1p(
    oil_rents["Oil_Rents_GDP"]
)

oil_rents.head()
```

```python
oil_rents.to_csv(
    INTERIM_DATA + "wdi_oil_rents_clean.csv",
    index=False
)
```

## GDP Growth

```python
gdp_growth = clean_wdi(
    RAW_DATA + "WDI_GDP_Growth_raw.xls",
    valid_codes,
    "GDP_growth"
)

gdp_growth.head()
```

```python
gdp_growth.to_csv(
    INTERIM_DATA + "wdi_gdp_growth_clean.csv",
    index=False
)
```

## GDP per Capita

```python
gdp_pc = clean_wdi(
    RAW_DATA + "WDI_GDP_perCapita_raw.xls",
    valid_codes,
    "GDP_per_capita"
)

gdp_pc["log_gdp_per_capita"] = np.log1p(
    gdp_pc["GDP_per_capita"]
)

gdp_pc.head()
```

```python
gdp_pc.to_csv(
    INTERIM_DATA + "wdi_gdp_per_capita_clean.csv",
    index=False
)
```

## Inflation (Winsorized)
Extreme values are capped at the 1st and 99th percentiles.

```python
inflation = clean_wdi(
    RAW_DATA + "WDI_Inflation_raw.xls",
    valid_codes,
    "Inflation"
)

# Winsorize at 1% and 99%
lower = inflation["Inflation"].quantile(0.01)
upper = inflation["Inflation"].quantile(0.99)

inflation["Inflation_w"] = inflation["Inflation"].clip(lower, upper)
```

```python
inflation.to_csv(
    INTERIM_DATA + "wdi_inflation_clean.csv",
    index=False
)
```

## Government Debt (% GDP)

```python
debt = clean_wdi(
    RAW_DATA + "WDI_Government_Debt_raw.xls",
    valid_codes,
    "Gov_Debt_GDP"
)

debt["log_debt"] = np.log1p(debt["Gov_Debt_GDP"])

debt.head()
```

```python
debt.to_csv(
    INTERIM_DATA + "wdi_debt_clean.csv",
    index=False
)
```
